```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%reload_ext google.colab.data_table
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
countries = tables[1]
countries.columns =['country', '1990', '2005', '2017', '2017(%world)', '2017vs1990', 'tCO2/km2/yr', 'tCO2/cap/yr', 'TotalincLUCF', 'TotalexcLUCF']
new_countries = countries[~countries["country"].str.contains("World|European", na=False)]

new_countries = new_countries.loc[:, "country":"2017"]
new_countries["relative change 1990"] = new_countries["1990"]/new_countries["1990"]*100
new_countries["relative change 2005"] = new_countries["2005"]/new_countries["1990"]*100
new_countries["relative change 2017"] = new_countries["2017"]/new_countries["1990"]*100
new_countries = new_countries.sort_values(by=['relative change 2017'])
new_countries = new_countries.reset_index(drop = True)
print(new_countries)
b3 = new_countries.loc[[0, 1, 2]]
print(b3)
t3 = new_countries[::-1][0:3]
t3 = t3.reset_index(drop = True)
print(t3)
fig = plt.figure()
fig, ax = plt.subplots()
x = ["1990", "2005", "2017"]
y1 = t3.loc[0,["relative change 1990", "relative change 2005", "relative change 2017"]]
y2 = t3.loc[1,["relative change 1990", "relative change 2005", "relative change 2017"]]
y3 = t3.loc[2,["relative change 1990", "relative change 2005", "relative change 2017"]]
y4 = b3.loc[0,["relative change 1990", "relative change 2005", "relative change 2017"]]
y5 = b3.loc[1,["relative change 1990", "relative change 2005", "relative change 2017"]]
y6 = b3.loc[2,["relative change 1990", "relative change 2005", "relative change 2017"]]


ax.plot(x, y1, label=t3.loc[0, "country"])
ax.plot(x, y2, label=t3.loc[1, "country"])
ax.plot(x, y3, label=t3.loc[2, "country"])
ax.plot(x, y4, label=b3.loc[0, "country"])
ax.plot(x, y5, label=b3.loc[1, "country"])
ax.plot(x, y6, label=b3.loc[2, "country"])
ax.legend()

ax.set_title("Graph 2: worst and best changers")
ax.grid(True)
ax.set_xlabel("year")
ax.set_ylabel("relative amount of CO2 emitted (1990 = 100%)")
```
